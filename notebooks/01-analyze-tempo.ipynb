```python
# 🔧 Setup
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# 🔊 Load your audio file (replace with your own path)
file_path1 = '../audio/test_clip_1.mp3'
file_path2 = '../audio/test_clip_2.mp3'
  # You can change this to any short mp3
y1, sr1 = librosa.load(file_path1)
y2, sr2 = librosa.load(file_path2)

# 🧠 Estimate tempo and beats
tempo1, beats1 = librosa.beat.beat_track(y=y1, sr=sr1)
tempo2, beats2 = librosa.beat.beat_track(y=y2, sr=sr2)
beat_times1 = librosa.frames_to_time(beats1, sr=sr1)
beat_times2 = librosa.frames_to_time(beats2, sr=sr2)

print(f"Estimated Tempo: {float(tempo1):.2f} BPM")
print(f"Beat count: {len(beat_times1)}")

# 📊 Visualize the waveform + beats
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y1, sr=sr1, alpha=0.6)
plt.vlines(beat_times1, ymin=-1, ymax=1, color='red')
plt.title('Detected Beat Positions')
plt.show()

print(f"Estimated Tempo: {float(tempo2):.2f} BPM")
print(f"Beat count: {len(beat_times2)}")

plt.figure(figsize=(14, 5))
librosa.display.waveshow(y2, sr=sr2, alpha=0.6)
plt.vlines(beat_times2, ymin=-1, ymax=1, color='red')
plt.title('Detected Beat Positions')
plt.show()
```
