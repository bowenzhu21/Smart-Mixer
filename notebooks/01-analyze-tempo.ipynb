```python
#  Setup
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

#  Load your audio file (replace with your own path)
file_path = '../audio/test_clip_1.mp3'  # You can change this to any short mp3
y, sr = librosa.load(file_path)

#  Estimate tempo and beats
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
beat_times = librosa.frames_to_time(beat_frames, sr=sr)

print(f"Estimated Tempo: {float(tempo):.2f} BPM")
print(f"Beat count: {len(beat_times)}")

#  Visualize the waveform + beats
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr, alpha=0.6)
plt.vlines(beat_times, ymin=-1, ymax=1, color='red')
plt.title('Detected Beat Positions')
plt.show()
```
